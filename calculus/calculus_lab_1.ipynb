Воспользуйтесь формулой вычисления будущей стоимости:
$$

В решении нужно использовать `solve`, `equation`

**<font color=#FF5EDD>Задача 10**</font> (3 балла)

Предположим, у Вас есть начальная сумма денег `initial`, и Вы хотите инвестировать ее под фиксированный годовой процентный доход `extra`. Вам нужно найти, сколько времени `time` потребуется, чтобы Ваша инвестиция достигла или превысила определенную целевую сумму `final`.

 Уравнение для этой задачи выглядит следующим образом:

$$ initial × \left(1 + \frac{extra}{100}\right)^{time} = final $$

Напишите функцию `check_time(initial: float, extra: float, final: float) -> int` чтобы определить сколько времени понадобится чтобы достичь желаемой цели.

```python
def check_time(initial, extra, final) -> int:
    time = 0 # Вместо нуля запишите искомое количиство времени
    time = math.ceil(
        math.log(final / initial) /
        math.log(1 + extra / 100)
    )
    return time
```
